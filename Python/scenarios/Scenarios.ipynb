```python
from IPython.display import Image, display
import glob
import cv2
import matplotlib.pyplot as plt

scenarios = []


YOKOHAMA = 'yokohama'
TAKANAWA = 'takanawa_gateway_station'
HANSAPLATZ = 'hansaplatz'
DRESDENSTATION = 'dresden_station'
HAMBURGSTATION = 'hamburg_station'

# 240-degree panorama pictures
IMG_FILES = "../../VirtualShowroom/Assets/Resources/Panorama/*.jpg"
paths = sorted(glob.glob(IMG_FILES))
paths
```

```python
filenames = {}
scenarios = {}
for p in paths:
    name = p.split('/')[-1].split('_')[:2]
    name = '_'.join(name)
    name = name.replace('.jpg', '')  # Remove file extentions (".jpg")
    filenames[name] = p

filenames
```

```python
def imshow(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
```

## Save the scenarios to SQLite database

```python
import sqlite3

with sqlite3.connect('db/scenarios.db') as conn:
    cur = conn.cursor()
    cur.execute('CREATE TABLE IF NOT EXISTS scenarios (image TEXT PRIMARY KEY, scene_id TEXT, scenario TEXT)')
    cur.execute('DELETE FROM scenarios')
    data = list(zip(scenarios.keys(), [e[1] for e in scenarios.values()], [e[0] for e in scenarios.values()]))
    cur.executemany('insert into scenarios(image, scene_id, scenario) values (?,?,?)', data)
```

```python
with sqlite3.connect('db/scenarios.db') as conn:
    cur = conn.cursor()
    records = cur.execute('SELECT * FROM scenarios').fetchall()

records
```
